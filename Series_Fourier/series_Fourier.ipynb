### __1.__ Исследовать сходимость ряда. 


### $$\frac{1}{2\sqrt2}+\frac{1}{3\sqrt3}+\,...\,+\frac{1}{(n+1)\sqrt{n+1}}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{1}{(n+1)\sqrt{n+1}}$$

(*) Двумя различными признаками.

### Решение:
### $$\frac{1}{2\sqrt2}+\frac{1}{3\sqrt3}+\,...\,+\frac{1}{(n+1)\sqrt{n+1}}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{1}{(n+1)\sqrt{n+1}}=\sum\limits_{n=1}^{+\infty}\frac{1}{(n+1)^{\frac{3}{2}}}$$
#### Небходимое условие:
### $$\lim_{n\to +\infty}\frac{1}{(n+1)^{\frac{3}{2}}}=0$$
#### Достаточное условие:
### $$\sum\limits_{n=1}^{+\infty}\frac{1}{(n+1)^{\frac{3}{2}}},\,\,\,\,\,\sim\,\,\,\,\,\,O\Bigr(\frac{1}{n^{\frac{3}{2}}}\Bigl)$$
#### $$т.к.\ степень\ равна \frac{3}{2}>1,\ то\ ряд\ сходится.$$

#### Теперь докажем сходимость используя интегральный признак Коши.

### $$\int\limits_{1}^{+\infty}a_{x}\,dx=\int\limits_{1}^{+\infty}\frac{1}{(n+1)\sqrt{n+1}} = \int\limits_{1}^{+\infty}(n+1)^{-\frac{3}{2}} = -2(n+1)^{-\frac{1}{2}}\bigg|_1^{+ \infty} = 0 + \frac{2}{\sqrt 2} =\sqrt 2$$
#### Так как несбственный интеграл существует и конечный, то ряд $\sum\limits_{n=1}^{+\infty}\frac{1}{(n+1)\sqrt{n+1}}$ сходится 

### __2.__ Исследовать сходимость ряда

### $$\frac{1000}{1!}+\frac{1000^2}{2!}+\,...\,+\frac{1000^n}{n!}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{1000^n}{n!}$$

### Решение:
#### воспльзуемся признаком д’Аламбера.

Найдем предел:

### $$\lim_{n\to +\infty}\frac{a_{n+1}}{a_{n}}=\lim_{n\to +\infty}\frac{\frac{1000^{n+1}}{n+1!}}{\frac{1000^n}{n!}}=\lim_{n\to +\infty}\frac{1000}{n+1} = 0,\ 0<1\ значит\ ряд\ сходится$$

### __3*.__ Исследовать сходимость ряда

### $$\frac{2\cdot1!}{1}+\frac{2^2\cdot2!}{2^2}+\,...\,+\frac{2^n\cdot n!}{n^n}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{2^n\cdot n!}{n^n}$$

### Решение:
#### воспльзуемся признаком д’Аламбера.

Найдем предел:

### $$\lim_{n\to +\infty}\frac{a_{n+1}}{a_{n}}=\lim_{n\to +\infty}\frac{\frac{2^{n+1}\cdot (n+1)!}{(n+1)^{n+1}}}{\frac{2^n\cdot n!}{n^n}} =\lim_{n\to +\infty}\frac{2\cdot n^n}{(n+1)^n} = 2\lim_{n\to +\infty}\frac{n^n}{(n+1)^n} =$$
### $$=2\lim_{n\to +\infty}e^{\ln n^n - \ln (n+1)^n} = 2\lim_{n\to +\infty}e^{n(\ln n - \ln (n+1))} = 2\cdot e^{\lim_{n\to +\infty}{n(\ln n - \ln (n+1))}}= 2\cdot e^{\lim_{n\to +\infty}\frac{(\ln n - \ln (n+1))}{1/n}}$$
#### по правилу Лопиталя
### $$2\cdot e^{\lim_{n\to +\infty}\frac{(\ln n - \ln (n+1))}{1/n}}=2e^{\lim_{n\to +\infty}\frac{\frac{1}{x} - \frac{1}{x+1}}{-\frac{1}{x^2}}} = 2e^{\lim_{n\to +\infty}-\frac{x}{x+1}} = 2\cdot e^{-1} < 1,\ ряд\ сходится$$


### __4*.__ Исследовать сходимость ряда

### $$\frac{3\cdot1!}{1}+\frac{3^2\cdot2!}{2^2}+\,...\,+\frac{3^n\cdot n!}{n^n}+\,...\,=\sum\limits_{n=1}^{+\infty}\frac{3^n\cdot n!}{n^n}$$

### Решение:
#### воспльзуемся признаком д’Аламбера.

Найдем предел:

### $$\lim_{n\to +\infty}\frac{a_{n+1}}{a_{n}}=\lim_{n\to +\infty}\frac{\frac{3^{n+1}\cdot (n+1)!}{(n+1)^{n+1}}}{\frac{3^n\cdot n!}{n^n}} =\lim_{n\to +\infty}\frac{3\cdot n^n}{(n+1)^n} = 3\lim_{n\to +\infty}\frac{n^n}{(n+1)^n} =$$
### $$=3\lim_{n\to +\infty}e^{\ln n^n - \ln (n+1)^n} = 3\lim_{n\to +\infty}e^{n(\ln n - \ln (n+1))} = 3\cdot e^{\lim_{n\to +\infty}{n(\ln n - \ln (n+1))}}= 3\cdot e^{\lim_{n\to +\infty}\frac{(\ln n - \ln (n+1))}{1/n}}$$
#### по правилу Лопиталя
### $$3\cdot e^{\lim_{n\to +\infty}\frac{(\ln n - \ln (n+1))}{1/n}}=3e^{\lim_{n\to +\infty}\frac{\frac{1}{x} - \frac{1}{x+1}}{-\frac{1}{x^2}}} = 3e^{\lim_{n\to +\infty}-\frac{x}{x+1}} = 3\cdot e^{-1} > 1,\ ряд\ расходится$$

### __6*.__ Разложить функцию $y=e^x$ в ряд Маклорена, а так же в ряд Фурье на отрезке $[-\pi, \pi]$. Построить график функции и разложений.

### Решение:
__Ряд Маклорена__ - это частный случай ряд Тейлора при $a=0$

### $$f(x)=\sum\limits_{n=0}^{+\infty}\frac{f^{(n)}(0)}{n!}x^n$$
### $$f^{(n)}=e^x,\ f^{(n)}(0) = 1$$
### $$e^x=1+\frac{1\cdot x}{1!}+\frac{1\cdot x^2}{2!}+\frac{1\cdot x^3}{3!}+ ... + \frac{1\cdot x^n}{n!}=\sum\limits_{n=0}^{+\infty}\frac{x^n}{n!}$$

```python
import math
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = "svg"

func_exp = []

i = 1
while i < 40:
    func_exp.append((math.e) ** i)
    i += 1
    
    
count_elem = 60
func_macloren = []
func_macloren_sum = []

for x in range(1, 40):
    for j in range(count_elem):
        func_macloren.append((x ** j) / math.factorial(j))
        
    func_macloren_sum.append(sum(func_macloren))
    func_macloren = []
    

fig,ax = plt.subplots(nrows=1,ncols=2)
ax1, ax2 = ax
ax1.plot(func_macloren_sum)
ax1.set_title('Ряд Маклорена($e^x$)')
ax2.plot(func_exp)
ax2.set_title('$e^x$')
```

#### Разложим в ряд Фурье на $[-\pi, \pi]$:

### $$f(x)=a_{0}+\sum\limits_{n=1}^{+\infty}(a_n\cos nx+b_n\sin nx)$$

где

### $$a_0=\frac{1}{2\pi}\int\limits_{-\pi}^\pi f(x)\,dx$$

### $$a_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\cos nx\,dx$$

### $$b_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\sin nx\,dx$$

### $$a_0=\frac{1}{2\pi}\int\limits_{-\pi}^\pi e^x\,dx = \frac{e^x}{2\pi} \bigg|_{-\pi}^{\pi} = \frac{e^{\pi} - e^{-\pi}}{2\pi}$$
### $$a_n =\frac{1}{\pi}\int\limits_{-\pi}^\pi e^{x}\cos nx\,dx$$ 

### $$U=\cos nx\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=-n\sin nx$$ 
### $$dV=e^xdx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=e^x$$ 

Тогда:

### $$a_n =\frac{1}{\pi}\int\limits_{-\pi}^\pi e^{x}\cos nx\,dx=\frac{e^{x}\cos nx}{\pi}+\frac{n}{\pi}\int\limits_{-\pi}^\pi e^{x}\sin nx\,dx$$ 

Воспользуемся формулой ещё раз

### $$U=\sin nx\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=n\cos nx$$ 
### $$dV=e^xdx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=e^x$$ 

Тогда:

### $$a_n =\frac{1}{\pi}\int\limits_{-\pi}^\pi e^{x}\cos nx\,dx=\frac{e^{x}\cos nx}{\pi}+\frac{ne^{x}\sin nx}{\pi}-\frac{n^2}{\pi}\int\limits_{-\pi}^\pi e^{x}\cos nx\,dx= \frac{e^x\cos nx+ne^x\sin nx}{n^2+1}\bigg|_{-\pi}^{\pi}=$$
### $$=\frac{e^{-\pi}-e^{\pi}}{n^2+1}$$ 
### $$b_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi e^x\sin nx\,dx$$
### $$U=\sin nx\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=n\cos nx$$ 
### $$dV=e^xdx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=e^x$$ 

Тогда:

### $$b_n =\frac{1}{\pi}\int\limits_{-\pi}^\pi e^{x}\sin nx\,dx=\frac{e^{x}\sin nx}{\pi}-\frac{n}{\pi}\int\limits_{-\pi}^\pi e^{x}\cos nx\,dx$$ 

Воспользуемся формулой ещё раз

### $$U=\cos nx\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=-n\sin nx$$ 
### $$dV=e^xdx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=e^x$$ 

Тогда:

### $$b_n =\frac{1}{\pi}\int\limits_{-\pi}^\pi e^{x}\cos nx\,dx=\frac{e^{x}\sin nx}{\pi}-\frac{ne^{x}\cos nx}{\pi}-\frac{n^2}{\pi}\int\limits_{-\pi}^\pi e^{x}\sin nx\,dx= \frac{e^x\sin nx-ne^x\cos nx}{n^2+1}\bigg|_{-\pi}^{\pi}=$$
### $$=\frac{n(e^{\pi}-e^{-\pi})}{n^2+1}$$ 

### $$e^x=\frac{e^{\pi} - e^{-\pi}}{2\pi}+\sum\limits_{n=1}^{+\infty}\frac{(e^{-\pi}-e^{\pi})(\cos nx - n\sin nx)}{n^2+1}$$
